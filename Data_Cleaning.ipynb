# imports

```python
import pandas as pd
import numpy as np
```

# load datasets

```python
transactions_df = pd.read_csv('csv_files/transactions_data_south_africa.csv')
mcc_code_df = pd.read_csv('csv_files/mcc_codes.csv')
fraud_labels_df =  pd.read_csv('csv_files/train_fraud_labels.csv')
user_df =  pd.read_csv('csv_files/user_data_south_africa.csv')
cards_df = pd.read_csv('csv_files/cards_data_south_africa.csv')
```

Display loaded data

```python
display(transactions_df.head(), 
cards_df.head(),
mcc_code_df.head(),
user_df.head(),
fraud_labels_df.head())
```

Column name standardization

```python
transactions_df = transactions_df.rename(columns={"id": "transaction_id"})
cards_df = cards_df.rename(columns={"id": "card_id"})
mcc_code_df = mcc_code_df.rename(columns={"mcc_code": "mcc"})
user_df = user_df.rename(columns={"id": "client_id"})
```

Displaying The standardized dataset

```python
display(transactions_df.head(), 
cards_df.head(),
mcc_code_df.head(),
user_df.head(),
fraud_labels_df.head())
```

```python
# mcc in transcations_df corresponds to mcc_code in mcc_code_df

mcc_code_df['mcc'] = mcc_code_df['mcc'].astype(float)
mcc_code_df
```

# Merging

```python
df = pd.merge(transactions_df, cards_df, how='left', on=['card_id', 'client_id'])
```

```python
df = df.merge(mcc_code_df, how="left", on='mcc')
```

```python
df = df.merge(user_df, how='left', on='client_id')
```

```python
df = df.merge(fraud_labels_df, how="left", on='transaction_id')
```

Basic Preprocessing


```python
def basic_preprocessing(df):
    df = df.copy()
    # full date and time
    df['date'] = pd.to_datetime(df['date'], format="%Y-%m-%d %H:%M:%S")
    # month and year only
    df['expires'] = pd.to_datetime(df['expires'])
    # year, month, day only
    df['acct_open_date'] = pd.to_datetime(df['acct_open_date'])
    # fill missing values
    df['errors'] = df['errors'].fillna('no error')
    df['mcc'] = df['mcc'].fillna(5812)
    df['description'] = df['description'].fillna("Eating Places, Restaurants")
    df['merchant_state'] = df['merchant_state'].fillna("Unknown")
    df['zip'] = df['zip'].fillna(0)
    # absolute value for amount
    df['neg_amount'] = (df['amount']<0).astype(int)
    df['amount_sign'] = np.sign(df['amount'])
    df['amount_abs'] = df['amount'].abs()
    return df

df = basic_preprocessing(df)
df
```
